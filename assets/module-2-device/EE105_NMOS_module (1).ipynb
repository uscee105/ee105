# Assignment 1 : Plot I<sub>D</sub>-V<sub>GS</sub> curve of the NMOS for V<sub>DS</sub>=3.3V

In this task, we'll measure the drain current I<sub>D</sub> as a function of the gate-source voltage V<sub>GS</sub> for an N-channel MOSFET (NMOS), and plot the transfer characteristic curve of I<sub>D</sub> vs. V<sub>GS</sub> while keeping the drain voltage V<sub>DS</sub>=3.3V.

```python
import serial
import time
import matplotlib.pyplot as plt
import csv

# Serial Port Configuration
SERIAL_PORT = "COM5"  # Update based on your system
BAUD_RATE = 115200  

# Constants
DAC_CONVERSION = 0.00165  # DAC step to voltage conversion

# Data storage for plotting
Vds_voltages = []  # NMOS Drain Voltage 
NMOS_currents = []  # Current through the NMOS
Vgs_voltages = []  # NMOS gate Voltage

def apply_voltage_and_read():
    """
    - Sends voltage steps to Arduino DAC to change Vgs
    - Reads ADC values for NMOS drain current
    - Stores data for plotting I-V curve
    """
    try:
        with serial.Serial(SERIAL_PORT, BAUD_RATE, timeout=3) as ser:
            time.sleep(2)  # Allow time for serial connection
            
            print("\nApplying voltage and reading output:\n")
            print(f"{'Vgs Voltage (V)':<25} {'Vds Voltage (V)':<25} {'NMOS Current (A)':<25}")
            print("="*75)

            ser.write(b"set_Vds 2000\n") #set Vds to 3.3V
            response = ser.readline().decode("utf-8").strip()
            time.sleep(0.05)
            
            inc = 0  # Start from 0V
            while inc <= 2000:  # Vgs max value for 3.3V
                # Send command to set Vgs voltage
                set_command = f"set_Vgs {inc}\n"
                ser.write(set_command.encode())  
                time.sleep(0.1)  # Allow voltage to settle

                # Read confirmation response
                response = ser.readline().decode("utf-8").strip()
                if not response:
                    print("⚠ Warning: No response from Arduino.")
                    continue

               # print(f"Arduino Response: {response}")  # Debugging: Show what Arduino sends

                # Request ADC reading
                ser.write(b"read_Id\n")
                time.sleep(0.1)

                # Read ADC response
                response = ser.readline().decode("utf-8").strip()
                if not response:
                    print("⚠ Warning: No ADC response received.")
                    continue

                #print(f"Raw ADC Response: '{response}'")  # Debugging: Show raw Arduino response

                try:
                    # Extract the drain current of NMOS from ADC response
                    NMOS_current = float(response.strip())  
                    Vds_voltage =3.3
                    Vgs_voltage= inc*DAC_CONVERSION;
                    
                    # Store data for plotting
                    Vgs_voltages.append(Vgs_voltage)
                    NMOS_currents.append(NMOS_current)
                    Vds_voltages.append(Vds_voltage)

                    # Print the values
                    print(f"{Vgs_voltage:.6f} V, {Vds_voltage:.6f} V, {NMOS_current:.9f} A")

                except ValueError:
                    print(f"⚠ Error: Unexpected response format: {response}")

                inc += 50  # Increase voltage step
            with open('NMOS_VGS_data.csv', mode='w') as file:
                writer = csv.writer(file, delimiter=",", lineterminator="\n")
                l=[Vgs_voltages,Vds_voltages,NMOS_currents]
                data=zip(*l)
                writer.writerows(data)
    
    
            print("\nMeasurement complete. DAC reset to 0V.")

            # Reset DAC to 0V
            ser.write(b"set_Vgs 0\n")
            time.sleep(0.05)
            ser.write(b"set_Vds 0\n")
            time.sleep(0.05)

    except serial.SerialException as e:
        print(f"Error: {e}")

def plot_iv_curve():
    """
    Plots the I-V curve (NMOS Current vs. Gate Voltage) without log scale
    """
    if not Vds_voltages or not NMOS_currents:
        print("⚠ Error: No valid data to plot!")
        return

    plt.figure(figsize=(8, 6))
    plt.plot(Vgs_voltages, NMOS_currents, marker='o', linestyle='-', color='r', label="NMOS I-Vgs Curve")
    
    plt.xlabel("Gate Voltage Vgs (V)")
    plt.ylabel("Drain Current Id(A)")
    plt.title("NMOS I-Vgs Characteristics")
    plt.legend()
    plt.grid(True)
    plt.show()

if __name__ == "__main__":
    apply_voltage_and_read()
    plot_iv_curve()  # Call the function to plot the I-V curve
```
