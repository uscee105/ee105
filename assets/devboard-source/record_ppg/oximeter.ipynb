```python
# open file

import os
# %cd C:\Users\tasif\Dropbox\stanford-paper-mentaid-placement\figures
#os.chdir(r"C:\Users\tasif\Dropbox\stanford-project-mental-health\WearM\stanford-project-mental-health-share\data")  
#os.chdir(r"G:\archive\archive\stanford\stanford-projects\stanford-project-niv\data\20200728")  # data not good

dir_name = %pwd


# import libs
import matplotlib
import os.path
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.ticker as ticker
from matplotlib import rcParams
from matplotlib import rc
import matplotlib.gridspec as gridspec

rc('font',**{'family':'sans-serif','sans-serif':['PT Sans'],'size': 35})
colors = ['#4E79A7', '#F28E2B', '#E15759', '#59A14E', '#EDC949', '#AF7AA1', '#FF9DA7', '#9C755F', '#BAB0AC', '#76B7B2']
from smooth_demo import smooth # smoothing function pulling from C:\Users\tasif\Dropbox\ipython

# open file 
from tkinter import filedialog
from tkinter import *

root = Tk()
fname1 = filedialog.askopenfilename(initialdir = dir_name,title = "Select file")
fname1_blank1 = os.path.basename(fname1)
fname1_blank1 = str(fname1).replace(".csv", "")
root.destroy() # get rid of the top-level instance once to make it actually invisible


#fname1 = "mouse_1_2020_8_22_8_57_55-1min-14hr.txt"
data1 = np.genfromtxt(fname1, delimiter=',', skip_footer=0, skip_header=1)
print (data1)
#time = np.linspace(0, 600, num=6000) 
```

```python
smoothing = 12
red_data = smooth(data1, smoothing)
nir_data = red_data
green_data = red_data
# green_data = smooth(data1[1], smoothing)
# nir_data = smooth(data1[2], smoothing)
# sc_data = smooth(data1[3], smoothing)
# cap_data = smooth(data1[4], smoothing)/1000

# for j in range(200):
#     sc_data[5300+j]=sc_data[5200]

# figure
fig, axs = plt.subplots(2, 1, figsize=(20,3*2),dpi=300, facecolor='w', edgecolor='w',gridspec_kw={'hspace': 1, 'wspace': 0.5})
# plot
axs[0].plot(red_data,linewidth=3, color=colors[0], alpha = 0.8, linestyle='--',
         label = "Battery voltage", marker="o",markersize=2)
axs[1].plot(red_data,linewidth=3, color=colors[1], alpha = 0.8, linestyle='--',
         label = "Ch 1 resistance", marker="o",markersize=2)
# axs[2].plot(nir_data,linewidth=3, color=colors[2], alpha = 0.8, linestyle='--',
#          label = "Ch 2 resistance", marker="o",markersize=2)
# axs[3].plot(sc_data,linewidth=3, color=colors[3], alpha = 0.8, linestyle='--',
#          label = "Ch 3 resistance", marker="o",markersize=2)
# axs[4].plot(cap_data,linewidth=3, color=colors[4], alpha = 0.8, linestyle='--',
#          label = "Ch 3 resistance", marker="o",markersize=2)

fig.savefig(str(fname1_blank1) +'_full.png', dpi=300, facecolor='w', edgecolor='w',
        orientation='portrait', papertype=None, format=None,
        transparent=False, bbox_inches='tight', pad_inches=.1)
```

```python
from scipy import signal
from scipy.signal import find_peaks
red_dc = max(red_data)
green_dc = max(green_data)
nir_dc = max(nir_data)

red_data = signal.detrend(red_data, type = 'linear')+red_dc
nir_data = signal.detrend(nir_data, type = 'linear')+nir_dc


# figure
fig, axs = plt.subplots(6, 1, figsize=(20,3*6),dpi=300, facecolor='w', sharex='col', edgecolor='w',gridspec_kw={'hspace': 1, 'wspace': 0.5})
start = 2000#yk1 2000
end = 2800#yk1 3000
time_step = .01
time = np.linspace(0, (end-start)*time_step, num=len(red_data[start:end])) 

#peak detection and heart rate
#dx = np.diff(-ys1) # differentiating the signal ys1 - rd, ys2 - ir
dx = nir_data[start:end]
# from peakdetect import peakdet
# maxtab, mintab = peakdet(dx,(max(dx)-min(dx))/2) # maxtab[:,0] given indexes of the peak and maxtab[:,1] give the values
# 5 for good one 
maxtab, _ = find_peaks(dx, prominence=1)      # BEST!
# # finding out bpm from the differences among the peaks
time_diff = np.diff(maxtab[1:-1,0])

heart_rate_bpm = 60.0/(time_diff/100.0)
#print "HR: "+str(np.around(heart_rate_bpm,decimals=2))+"b.p.m."

rmssd_array = [];
for i in range(0, len(time_diff)):
    print(time_diff[i:i+4])
    rmssd_array.append(np.sqrt(np.mean(np.square(time_diff[i:i+3]))))
print (rmssd_array)
print (len(time_diff))

for i in range(len(heart_rate_bpm)):
    if (np.abs(heart_rate_bpm[i]-heart_rate_bpm[i-1])>10):
        heart_rate_bpm[i]=heart_rate_bpm[i-1]


# plot
axs[0].plot(time, red_data[start:end],linewidth=3, color="red", alpha = 0.6, linestyle='-',
         label = "PPG (Red)")#, marker="o",markersize=2)
# axs[1].plot(time, green_data[start:end],linewidth=3, color="green", alpha = 0.6, linestyle='-',
#          label = "PPG (Green)")#, marker="o",markersize=2)
axs[1].plot(time, nir_data[start:end],linewidth=3, color="grey", alpha = 0.6, linestyle='-',
         label = "PPG (NIR)")#, marker="o",markersize=2)
axs[2].plot(time,dx,linewidth=2)
axs[2].plot((mintab[1:-1,0])/100.0, mintab[1:-1,1],'ro',linewidth=3)
axs[2].plot((maxtab[1:-1,0])/100.0, maxtab[1:-1,1],'bo',linewidth=3)
#axs[4].plot(time,dx,linewidth=0)
axs[3].plot((maxtab[2:-1,0])/100.0, heart_rate_bpm,'ko-',linewidth=3, color=colors[1])
#axs[5].plot(time,dx,linewidth=0)
axs[4].plot((maxtab[2:-1,0])/100.0, rmssd_array,'ko-',linewidth=3, color=colors[2])
axs[5].plot(time, sc_data[start:end],linewidth=3, color=colors[3], alpha = 0.6, linestyle='-',
         label = "Skin conductance")#, marker="o",markersize=2)
# axs[4].plot(time, cap_data[start:end],linewidth=3, color=colors[3], alpha = 0.8, linestyle='--',
#          label = "Ch 3 resistance", marker="o",markersize=2)
# labels and legends

for ax in axs.reshape(-1): 
    #ax.set_ylabel('$\Delta C$ (pF)')
    ax.legend(loc='upper center', bbox_to_anchor=(0.8, 1.7), ncol=2, fontsize = 30, fancybox=True,frameon=False)# bbox_to_anchor=(1, 1)#\
    #ax.legend.get_frame().set_alpha(0)
    ax.grid(b=True, which='major', color='k', linestyle='--', alpha = 0.2)
    ax.grid(b=True, which='minor', color='k', linestyle='--', alpha = 0.2)
axs[5].set_xlabel('Time (s)')
axs[0].set_ylabel('PPG (V)')
axs[1].set_ylabel('PPG (V)')
axs[2].set_ylabel('PPG (V)')
axs[5].set_ylabel('SC ($\mu$S)')
axs[3].set_ylabel('PPG peaks')

axs[3].set_ylabel('HR \n(b.p.m.)')
axs[3].yaxis.set_ticks(np.linspace(50,100, num = 4))
axs[3].yaxis.set_major_formatter(ticker.FormatStrFormatter('%2.1f'))

text_hr = r"$\overline{HR}$= "+str(np.around(np.mean(heart_rate_bpm),decimals=2))+"b.p.m."
plt.text(.15, .18,text_hr, horizontalalignment='center', verticalalignment='center',transform = axs[3].transAxes)

axs[4].set_ylabel('HRV \n(ms)')
axs[4].yaxis.set_ticks(np.linspace(50,100, num = 4))
axs[4].yaxis.set_major_formatter(ticker.FormatStrFormatter('%2.1f'))


plt.tight_layout()

text_hrv = r"$\overline{HRV}$= "+str(np.around(np.sqrt(np.mean(np.square(time_diff))),decimals=2))+"ms"
plt.text(.15, .18,text_hrv, horizontalalignment='center', verticalalignment='center',transform = axs[4].transAxes)
# # gridlines
# rcParams['xtick.direction'] = 'out'
# rcParams['ytick.direction'] = 'out'
text_sc = r"$\overline{SC}$= "+str(np.around(np.mean(sc_data[start:end]),decimals=2))+'$\mu$S'
plt.text(.15, .18,text_sc, horizontalalignment='center', verticalalignment='center',transform = axs[5].transAxes)
# export figure
fig.savefig(str(fname1_blank1) +'_zoom2.png', dpi=300, facecolor='w', edgecolor='w',
        orientation='portrait', papertype=None, format=None,
        transparent=False, bbox_inches='tight', pad_inches=.1)
```
